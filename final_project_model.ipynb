```python
# import packages
import pandas as pd
import numpy as np
```

```python
# load the dataset
from sklearn.model_selection import train_test_split

df = pd.read_csv('tainan_house.csv')
display(df.head())

X = df.iloc[:,:-1]
y = df.iloc[:,-1]

# split the dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
# data normalization
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(X_train.values)  # learn distribution of traning data
X_train_nor = sc.transform(X_train.values) # 學完的data
X_test_nor = sc.transform(X_test.values)
```

```python
# simple linear regression
from sklearn.linear_model import LinearRegression
regr = LinearRegression()
regr.fit(X_train_nor, y_train.values)
r_squared = regr.score(X_train_nor, y_train.values)

print('截距項', regr.intercept_)
print('X係數', regr.coef_) # f(x)
print('R平方:',r_squared)
```
