```python
from sklearn import tree
from sklearn import datasets
import pandas as pd
import pydotplus
df = pd.read_csv('110_整理.csv')
```

```python
# 
df = df.drop(labels=['單價每平方公尺'], axis = 1)
df = df.dropna(subset = ['總樓層數', '建物型態', '房', '衛', '廳'])
df = df.reset_index(drop=True)

df['車位移轉總面積平方公尺'] = df['車位移轉總面積平方公尺'].fillna(0)
df['電梯'] = df['電梯'].map({'無':0, '有':1})
df['有無管理'] = df['有無管理'].map({'無':0, '有':1})
df['建物型態'] = df['建物型態'].map({'透天厝':1, '住宅大樓':2, '華夏':3, '公寓':4})
df['土地總面積平方公尺'] = df['土地總面積平方公尺'].fillna(0)
df = df.dropna(subset = ['建物型態'])
```

```python
df = df.rename(columns = {'車位移轉總面積平方公尺': '車位'})
df['車位'] = df['車位'].apply(lambda x: 1 if x !=0 else x)
price = df['總價元']/10000
df = df.drop(labels=['總價元'], axis = 1)
```

```python
one_hot_encoding = pd.get_dummies(df['鄉鎮市區'], prefix = '鄉鎮市區')
df = pd.concat([df, one_hot_encoding, price], axis = 1)  # 合併資料
df = df.drop('鄉鎮市區', axis = 1)
df = df.rename(columns = {'總價元': '總價元(萬)'})
display(df)
```
